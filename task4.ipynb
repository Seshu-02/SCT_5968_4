```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample accident dataset
data = {
    'Start_Time': ['2021-06-01 08:00:00', '2021-06-01 17:30:00', '2021-06-02 23:15:00', 
                   '2021-06-03 03:45:00', '2021-06-03 12:00:00'],
    'Weather_Condition': ['Clear', 'Rain', 'Snow', 'Clear', 'Fog'],
    'Road_Condition': ['Dry', 'Wet', 'Snowy', 'Dry', 'Wet'],
    'Severity': [2, 3, 4, 2, 3],
    'City': ['Los Angeles', 'New York', 'Chicago', 'Houston', 'Phoenix'],
    'Latitude': [34.05, 40.71, 41.88, 29.76, 33.45],
    'Longitude': [-118.24, -74.00, -87.63, -95.37, -112.07]
}

# Create DataFrame
df = pd.DataFrame(data)

# Convert Start_Time to datetime and extract hour
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['Hour'] = df['Start_Time'].dt.hour

# Set seaborn style
sns.set(style="darkgrid")

# --- Plot 1: Accidents by Weather Condition ---
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Weather_Condition', hue='Weather_Condition', palette='muted', legend=False)
plt.title('Accidents by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.show()

# --- Plot 2: Accidents by Road Condition ---
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Road_Condition', hue='Road_Condition', palette='Set2', legend=False)
plt.title('Accidents by Road Condition')
plt.xlabel('Road Condition')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.show()

# --- Plot 3: Accidents by Hour of Day ---
plt.figure(figsize=(6, 4))
sns.histplot(df['Hour'], bins=6, kde=False, color='orange')
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.show()

# --- Plot 4: Heatmap of Accident Locations ---
plt.figure(figsize=(6, 5))
sns.kdeplot(x=df['Longitude'], y=df['Latitude'], cmap="Reds", fill=True, bw_adjust=1)
plt.scatter(df['Longitude'], df['Latitude'], c='blue', s=30, label='Accident Location')
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.tight_layout()
plt.show()

```
